```python
def find_template2d(piece, image):
    length_piece = piece.shape[0]
    width_piece = piece.shape[1]

    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            image_piece = image[i:length_piece+i, j:width_piece+j]
            if np.array_equal(piece, image_piece): 
                return image[i:length_piece+i, j:width_piece+j]
    return 'nope'
```

```python
found = find_template2d(piece, image)
```
